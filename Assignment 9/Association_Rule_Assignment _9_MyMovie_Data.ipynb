```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(10,6)
plt.rcParams['figure.dpi']=250
import warnings
warnings.filterwarnings('ignore')
from mlxtend.frequent_patterns import association_rules,apriori
```

# Movie Data

```python
movie=pd.read_csv("my_movies.csv")
movie
```

```python
mv=movie.drop(columns=['V1','V2','V3','V4','V5'], axis=1)
mv
```

```python
mv.info()
```

# Apriori algorithm

```python
frequent_movie = apriori(mv,min_support=0.1,use_colnames=True)
frequent_movie 
```

```python
#Building association rule
rules = association_rules(frequent_movie,metric='lift',min_threshold=0.8)
rules
```

```python
rule=rules.sort_values('lift',ascending = False)[0:20]
```

```python
rule
```

```python
rules[rules.lift>1]
```

```python
plot=rule.plot.bar(figsize=(25,8))
```

```python
def to_list(i):
    return(sorted(list(i)))
```

```python
ma_X = rules.antecedents.apply(to_list)+rules.consequents.apply(to_list)
```

```python
ma_X
```

```python
ma_X=ma_X.apply(sorted)
ma_X
```

```python
rules_sets = list(ma_X)
rules_sets
```

```python
unique_rules_sets = [list(m) for m in set(tuple(i) for i in rules_sets)]
index_rules=[]

for i in unique_rules_sets:
    index_rules.append(rules_sets.index(i))
    
unique_rules_sets  
```

```python
#getting rules without any redundancy
rules_no_redundancy= rules.iloc[index_rules,:]
```

```python
rules_no_redundancy.head(10)
```

```python
#shorting them w.r.t. list and getting top 5 rules
rules_movie = rules_no_redundancy.sort_values('lift',ascending=False).head()
```

```python
rules_movie.head()
```
