# Apriori Algorithm

```python
frequent_books = apriori(book, min_support=0.1, use_colnames=True)
frequent_books
```

```python
rules = association_rules(frequent_books,metric='lift',min_threshold=0.6)
rules
```

```python
rule=rules.sort_values('lift',ascending = False)[0:20]
rule
```

```python
rules[rules.lift>1]
```

```python
plot = rule.plot.bar(figsize=(20,5))
```

```python
def to_list(i):
    return(sorted(list(i)))
```

```python
ma_X = rules.antecedents.apply(to_list)+rules.consequents.apply(to_list)
```

```python
ma_X
```

```python
ma_X=ma_X.apply(sorted)
ma_X
```

```python
rules_sets = list(ma_X)
rules_sets
```

```python
unique_rules_sets = [list(m) for m in set(tuple(i) for i in rules_sets)]
index_rules=[]

for i in unique_rules_sets:
    index_rules.append(rules_sets.index(i))
    
unique_rules_sets   
```

```python
#getting rules without any redundancy
rules_no_redundancy= rules.iloc[index_rules,:]
```

```python
rules_no_redundancy.head(10)
```

```python
#shorting them w.r.t. list and getting top 5 rules
rule_book = rules_no_redundancy.sort_values('lift',ascending=False).head()
```
