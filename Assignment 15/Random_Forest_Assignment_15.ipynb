```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn import tree
plt.rcParams["figure.figsize"]=(10,5)
plt.rcParams["figure.dpi"]=300
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

# Fraud Check Data

```python
FC=pd.read_csv("Fraud_check.csv")
FC
```

# EDA

```python
FC.info()
```

```python
FC.describe()
```

```python
FC=pd.get_dummies(FC)
FC
```

```python
FC['Taxable.Income'].max()
```

```python
FC['Taxable.Income'].min()
```

```python
FC['Tax']=pd.cut(FC['Taxable.Income'], bins = [10002,30000,99620], labels = ["Risky", "Good"])
```

```python
FC
```

```python
array=FC.values
array
```

```python
S=array[:,1:10]
T=array[:,10]
```

```python
S
```

```python
T
```

# Model Building

```python
kfold=KFold(n_splits=10,shuffle=True)
```

```python
model1=RandomForestClassifier(n_estimators=100,max_features=5)
model1
```

```python
fit1=model1.fit(S,T)
pred=fit1.predict(S)
pred
```

```python
result=cross_val_score(model1,S,T,cv=kfold)
result
```

```python
result.mean()*100
```
